```python
from PIL import Image, ImageDraw, ImageFont
import os

def add_text_to_images(image_folder, output_folder, font_path, font_size=40):
    """
    åœ¨æ¯å¼ å›¾ç‰‡çš„æ­£ä¸­é—´åä¸Š1/3çš„ä½ç½®æ·»åŠ å¯¹åº”çš„æ–‡æœ¬ã€‚
    
    :param image_folder: å›¾ç‰‡æ–‡ä»¶å¤¹è·¯å¾„
    :param text_list: æ–‡æœ¬åˆ—è¡¨ï¼Œä¸å›¾ç‰‡ä¸€ä¸€å¯¹åº”
    :param output_folder: è¾“å‡ºæ–‡ä»¶å¤¹è·¯å¾„
    :param font_path: å­—ä½“æ–‡ä»¶è·¯å¾„
    :param font_size: å­—ä½“å¤§å°
    """
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # è·å–å›¾ç‰‡æ–‡ä»¶åˆ—è¡¨
    image_files = sorted([f for f in os.listdir(image_folder) if f.lower().endswith(('png', 'jpg', 'jpeg'))])
    
    # if len(image_files) != len(text_list):
    #     raise ValueError("å›¾ç‰‡æ•°é‡å’Œæ–‡æœ¬æ•°é‡ä¸åŒ¹é…ï¼")
    
    for i, image_file in enumerate(image_files):
        image_path = os.path.join(image_folder, image_file)
        split_image_file = image_file[:-4].split("_")
        like_count = int(split_image_file[-1])
        if like_count < 1:
            continue
        # text = text_list[i]
        text = split_image_file[0]
        
        # æ‰“å¼€å›¾ç‰‡
        try:
            with Image.open(image_path) as img:
                draw = ImageDraw.Draw(img)
                width, height = img.size
                
                # åŠ è½½å­—ä½“
                try:
                    font = ImageFont.truetype(font_path, font_size)
                except IOError:
                    raise IOError(f"æ— æ³•åŠ è½½å­—ä½“æ–‡ä»¶ï¼š{font_path}")
                
                # è®¡ç®—æ–‡æœ¬ä½ç½®
                text_bbox = font.getbbox(text)  # è·å–æ–‡æœ¬çš„è¾¹ç•Œæ¡†
                text_width = text_bbox[2] - text_bbox[0]  # å®½åº¦
                text_height = text_bbox[3] - text_bbox[1]  # é«˜åº¦
                x = (width - text_width) // 2
                y = height * 46/50 # åä¸Š1/3
                
                # è®¾ç½®æ–‡æœ¬é¢œè‰²ï¼ˆè‰ºæœ¯çº¢è‰²ï¼‰
                text_color = (0, 255, 0)  # çº¢è‰²
                
                # æ·»åŠ æ–‡æœ¬åˆ°å›¾ç‰‡
                draw.text((x, y), text, fill=text_color, font=font)
                
                # ä¿å­˜å›¾ç‰‡
                output_path = os.path.join(output_folder, image_file)
                img.save(output_path)
                print(f"å·²å¤„ç†å›¾ç‰‡ï¼š{output_path}")
        except Exception as e:
            print(f"å¤„ç†å›¾ç‰‡æ—¶å‡ºé”™ï¼š{image_path}ï¼Œé”™è¯¯ä¿¡æ¯ï¼š{e}")
            continue

# ç¤ºä¾‹ç”¨æ³•
if __name__ == "__main__":
    image_folder = "C:\\Users\\yanmeizhao\\PycharmProjects\\CodeRepo\\crawling\\images_"  # æ›¿æ¢ä¸ºä½ çš„å›¾ç‰‡æ–‡ä»¶å¤¹è·¯å¾„
    text_list = ["æ–‡æœ¬1", "æ–‡æœ¬2", "æ–‡æœ¬3"]  # æ›¿æ¢ä¸ºä½ çš„æ–‡æœ¬åˆ—è¡¨
    output_folder = "C:\\Users\\yanmeizhao\\PycharmProjects\\CodeRepo\\crawling\\selected_with_text"  # æ›¿æ¢ä¸ºè¾“å‡ºæ–‡ä»¶å¤¹è·¯å¾„
    font_path = "C:\\Users\\yanmeizhao\\Downloads\\ZCOOL_KuaiLe\\ZCOOLKuaiLe-Regular.ttf"  # æ›¿æ¢ä¸ºä½ çš„å­—ä½“æ–‡ä»¶è·¯å¾„
    font_size = 50  # æ ¹æ®éœ€è¦è°ƒæ•´å­—ä½“å¤§å°
    
    add_text_to_images(image_folder, output_folder, font_path, font_size)
```
